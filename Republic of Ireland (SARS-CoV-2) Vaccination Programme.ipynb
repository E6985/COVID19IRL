### Republic of Ireland (SARS-CoV-2) Vaccination Programme
***

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

nAsOf="Sunday, 15th Aug 2021"
nSingleDose=69.1
nFullyVaccinated=64.7
nVulnerablePopulation=35.3

rcParams["text.color"]="k"
labels="Vulnerable","Vaccinated (inc. Janssen)"
sizes=[nVulnerablePopulation,nFullyVaccinated]
colors=["red","green"]
explode=(0.25,0) # explode 1st slice

plt.title("Republic of Ireland (SARS-CoV-2) Vaccination Programme")
plt.text(-1.35,1.0,s=nAsOf,horizontalalignment="center")

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=240)
plt.text(1.1,-1.3,"Single Dose {}%".format(nSingleDose),horizontalalignment="center")
plt.axis('equal')
plt.savefig(fname="../rc/covid-19-ireland/RepublicofIrelandSARS-CoV-2VaccinationProgramme.png")
```
